```python
df_country = pd.read_excel(r'D:\PROJECT\01-EDA\Zomato Data Analysis\Country-Code.xlsx')
```

```python
df = pd.merge(df_zomato, df_country, on = 'Country Code', how = 'left')
```

#### Perform exploratory data analysis and visuaization

```python
df.columns
```

```python
country_name = df.Country.value_counts().index
countr_val = df.Country.value_counts().values
```

```python
plt.pie(countr_val[:3], labels = country_name[:3], autopct = '%1.2f%%');
```

```python
rating = df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0:'Rating count'})
```

```python
plt.figure(figsize=(10, 3))
sns.barplot(data = rating, x = 'Aggregate rating', y = 'Rating count', hue = 'Rating color', palette = ['white', 'red', 'orange', 'yellow', 'green', 'green'])
```
